2. Reorganise Folders

```python
import os
import shutil

# Define the path to your primary folders
primary_folder_path = folders_path

# Loop over each primary folder
for primary_folder in os.listdir(primary_folder_path):
    primary_folder_full_path = os.path.join(primary_folder_path, primary_folder)
    
    # Check if the current item is a directory
    if os.path.isdir(primary_folder_full_path):
        # Loop over each secondary folder inside the primary folder
        for secondary_folder in os.listdir(primary_folder_full_path):
            secondary_folder_full_path = os.path.join(primary_folder_full_path, secondary_folder)
            
            # Check if the current item is a directory
            if os.path.isdir(secondary_folder_full_path):
                # Move the contents of the secondary folder one level up into the primary folder
                for item in os.listdir(secondary_folder_full_path):
                    item_full_path = os.path.join(secondary_folder_full_path, item)
                    shutil.move(item_full_path, primary_folder_full_path)
                
                # Remove the secondary folder
                shutil.rmtree(secondary_folder_full_path)

```

```python
import os
import shutil

# Define the parent directory containing the folders with similar names
parent_directory = folders_path

# Get a list of all items in the parent directory
items = os.listdir(parent_directory)

# Create a dictionary to store the folder names as keys and the folder paths as values
folders_dict = {}

# Iterate through the items in the parent directory
for item in items:
    # Get the full path of the item
    item_path = os.path.join(parent_directory, item)
    # Check if the item is a directory
    if os.path.isdir(item_path):
        # Extract the folder name without the last letter
        folder_name = item[:-1]
        # Append the folder path to the dictionary using the folder name as the key
        if folder_name not in folders_dict:
            folders_dict[folder_name] = []
        folders_dict[folder_name].append(item_path)

# Iterate through the folders in the dictionary and combine their data
for folder_name, folder_paths in folders_dict.items():
    # Check if there are multiple folders with the same name, except for the last letter
    if len(folder_paths) > 1:
        print(f"Combining data in folders with name '{folder_name}':")
        # Define the destination folder path to combine the data
        destination_folder = os.path.join(parent_directory, f"{folder_name}_combined")
        # Create the destination folder if it doesn't exist
        if not os.path.exists(destination_folder):
            os.mkdir(destination_folder)
        # Iterate through the folder paths and move the data to the destination folder
        for folder_path in folder_paths:
            # Get the files and folders inside the current folder
            items = os.listdir(folder_path)
            # Move the items to the destination folder
            for item in items:
                item_path = os.path.join(folder_path, item)
                destination_path = os.path.join(destination_folder, item)
                if os.path.isfile(item_path):
                    # If the item is a file, move it to the destination folder
                    shutil.move(item_path, destination_path)
                elif os.path.isdir(item_path):
                    # If the item is a directory, move its contents to the destination folder
                    for sub_item in os.listdir(item_path):
                        sub_item_path = os.path.join(item_path, sub_item)
                        sub_destination_path = os.path.join(destination_folder, sub_item)
                        shutil.move(sub_item_path, sub_destination_path)
                    # Remove the empty directory
                    os.rmdir(item_path)
            # Print the progress
            print(f" - Moved data from '{folder_path}' to '{destination_folder}'")
        # Remove the original folders
        for folder_path in folder_paths:
            os.rmdir(folder_path)
        print(" - Original folders removed")
```

```python
# 4. Data Reduction Function
def data_reduction_fun(df_eventfiles,df_properties,global_path,unique_ids = True, min_counts = 0):
    """
    DESCRIPTION: Reduces evenfiles table and properties table to required columns and adds unique ID, can now be used for data representation function
    INPUT: 1. Original eventfile table, 2. Original properties table, 3. Global Path, 4. Set Name
    OUTPUT: 1. Reduced eventfile table, 2. Reduced properties table
    """
    # Extract important labels and input columns
    df_eventfiles_input = df_eventfiles[['obsid','region_id','time','energy','chipx','chipy']]
    df_properties_input = df_properties[['name','obsid','region_id','hard_hm','hard_hs','hard_ms','var_prob_b','var_prob_h','var_prob_m','var_prob_s','var_index_b']]
    # Create unique IDs
    df_eventfiles_input['obsreg_id'] = df_eventfiles_input['obsid'].astype(str) + '-' + df_eventfiles_input['region_id'].astype(str)
    df_properties_input['obsreg_id'] = df_properties_input['obsid'].astype(str) + '-' + df_properties_input['region_id'].astype(str)
    # Drop individual IDs
    df_eventfiles_input = df_eventfiles_input.drop(columns=['obsid', 'region_id'])
    df_properties_input = df_properties_input.drop(columns=['obsid', 'region_id'])
    # Filter observations where there are less than min_counts counts
    counts = df_eventfiles_input['obsreg_id'].value_counts()
    count_mask = df_eventfiles_input['obsreg_id'].isin(counts[counts >= min_counts].index)
    df_eventfiles_input = df_eventfiles_input[count_mask]
    # Unique ID combinations filter
    if unique_ids:
        df_eventfiles_input = df_eventfiles_input[df_eventfiles_input['obsreg_id'].isin(df_properties_input['obsreg_id'].unique())]
        df_properties_input = df_properties_input[df_properties_input['obsreg_id'].isin(df_eventfiles_input['obsreg_id'].unique())]
        df_eventfiles_input = df_eventfiles_input[df_eventfiles_input['obsreg_id'].isin(df_properties_input['obsreg_id'].unique())]
    df_eventfiles_input = df_eventfiles_input.sort_values(by=['obsreg_id','time'])
    df_properties_input = df_properties_input.sort_values(by='obsreg_id')
    df_properties_input.insert(1, 'obsreg_id', df_properties_input.pop('obsreg_id'))
    # Save new dataframes
    df_eventfiles_input.to_csv(f'{global_path}/eventfiles-input-Full.csv',index=False)
    df_properties_input.to_csv(f'{global_path}/properties-input-Full.csv',index=False)
    return df_eventfiles_input, df_properties_input
```
